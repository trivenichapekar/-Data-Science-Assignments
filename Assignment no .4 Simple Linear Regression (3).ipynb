# Transformation for Continuous Variable 

```python
# log trasformation :
data['log_YearsExperience']= np.log(data["YearsExperience"])

fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].hist(data['YearsExperience']);ax[0].set_title("YearsExperience")
ax[1].hist(data['log_YearsExperience']);ax[1].set_title("Log_YearsExperience")
plt.show()
```

```python

data['log_Salary']= np.log(data["Salary"])

fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].hist(data['Salary']);ax[0].set_title("Salary")
ax[1].hist(data['log_Salary']);ax[1].set_title("Log_Salary")
plt.show()
```

```python

# square trasformation :
data['sqr_YearsExperience']= np.square(data["YearsExperience"])

fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].hist(data['YearsExperience']);ax[0].set_title("YearsExperience")
ax[1].hist(data['sqr_YearsExperience']);ax[1].set_title("Square_YearsExperience")
plt.show()
```

```python
data['sqr_Salary']= np.square(data["Salary"])

fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].hist(data['Salary']);ax[0].set_title("Salary")
ax[1].hist(data['sqr_Salary']);ax[1].set_title("Square_Salary")
plt.show()
```

```python
# square root trasformation :
data['sqrt_YearsExperience']= np.sqrt(data["YearsExperience"])

fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].hist(data['YearsExperience']);ax[0].set_title("YearsExperience")
ax[1].hist(data['sqrt_YearsExperience']);ax[1].set_title("Sqrt of YearsExperience")
plt.show()
```

```python

data['sqrt_Salary']= np.sqrt(data["Salary"])

fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].hist(data['Salary']);ax[0].set_title("Salary")
ax[1].hist(data['sqrt_Salary']);ax[1].set_title("Sqrt of Salary")
plt.show()
```
